```python
import pylab as pl
import csv
import numpy as np
#Conseguir la informacion del archivo CSV
entrada=open('T1_Housing.csv')
tabla=[]
for fila in csv.reader(entrada):
    tabla.append(fila)
entrada.close()
#Colocar la informacion obtenida en diferentes listas
x_inSample=[]
x_outSample=[]
y_inSample=[]
y_outSample=[]
for fila in range(1, len(tabla)):
    if tabla[fila][3]==('1'):
        x_inSample.append(float(tabla[fila][1]))
        y_inSample.append(float(tabla[fila][2]))
    else:
        x_outSample.append(float(tabla[fila][1]))
        y_outSample.append(float(tabla[fila][2]))
#Graficar información obtenida
pl.figure(figsize=(15,9))   
pl.plot(x_inSample,y_inSample,'*b',label='Training')
pl.plot(x_outSample,y_outSample,'*c',label='Validation')
pl.legend(loc='upper left', bbox_to_anchor=(0, 0.5, 0.5, 0.5)) 
pl.xlabel('Average Area Income')
pl.ylabel('Price')
pl.title('Datos Graficados')
pl.show()

```

```python
lista=np.arange(0,10,0.1)
#Funcion para agregar columnas de unos
def x_ne(A):
    z=[]
    for i in range(0,len(A)):
        z.append([A[i],1])
    return z
#Funcion que entrega parametros de la regresion lineal
def reg_lineal(X,Y,rho):
    X_ne=x_ne(X)
    X_netrans=np.transpose(X_ne)
    fPart=np.linalg.inv(np.dot(X_netrans,X_ne)+rho*np.identity(len(X_ne[0])))
    sPart=np.dot(X_netrans,Y)
    theta=np.dot(fPart,sPart)
    return theta

#Obtener diferentes valores del parametro según el valor de rho
a=[]
b=[]
for i in lista:
    a.append(reg_lineal(x_inSample,y_inSample,i)[0])
    b.append(reg_lineal(x_inSample,y_inSample,i)[1])

#Graficar diferentes valores del parametro según el valor de rho
pl.figure(figsize=(15,9))
pl.subplot(2,2,1)
pl.plot(lista,a,'b')
pl.xlabel(r"$\rho$")
pl.ylabel(r"Primer valor de $\Theta$")
pl.title(r"Primer valor del parametro $\Theta$ en función de $\rho$")

pl.subplot(2,2,2)
pl.plot(lista,b,'b')
pl.xlabel(r"$\rho$")
pl.ylabel(r"Segundo valor de $\Theta$")
pl.title(r"Segundo valor del parametro $\Theta$ en función de $\rho$")
pl.show()
```

```python
#Crear listas para error cuadratico y varianza
errorIn=[]
errorOut=[]
varIn=[]
varOut=[]

#Calcular el valor de Sigma para la varianza
def sigma(x,y,rho):
    n=len(x)
    f=(np.dot(x,reg_lineal(x,y,rho)[0]))+reg_lineal(x,y,rho)[1]
    s=0
    for i in range(0,n):
        s+=(y[i]-f[i])**2
    return s/n

#Calcular el valor de la varianza
def Varianza(x1,y1,x2,rho):
    a=0;
    Xt=x_ne(x1)
    Xv=x_ne(x2)
    Sig=sigma(x1,y1,rho)
    n=len(Xt[0])
    X2=np.dot(np.transpose(Xt),Xt)
    for i in range(0,len(x2)):
        F=np.dot(Xv[i],np.linalg.inv(X2+np.identity(n)*rho))
        S=np.dot(np.dot(X2,np.linalg.inv(X2+np.identity(n)*rho)),np.transpose(Xv[i]))  
        funcion=Sig*np.dot(F,S)
        a+=funcion
    return a/len(x2)

#Calcular el valor del error cuadratico
def MSE(xval,yval,xtrain,ytrain,rho):
    E=0
    n=len(xval)
    funcion=(np.dot(xval,reg_lineal(xtrain,ytrain,rho)[0]))+reg_lineal(xtrain,ytrain,rho)[1]
    for i in range(n):
        E+=(yval[i]-funcion[i])**2
    return E/n
    
    
#Completar listas para la varianza y el error cuadratico     
for a in lista:
    errorOut.append(MSE(x_outSample,y_outSample,x_inSample,y_inSample,a))
    errorIn.append(MSE(x_inSample,y_inSample,x_inSample,y_inSample,a))
    varIn.append(Varianza(x_inSample,y_inSample,x_inSample,a))
    varOut.append(Varianza(x_inSample,y_inSample,x_outSample,a))

#Graficar error cuadratico y varianza para diferentes valores de rho
pl.figure(figsize=(15,9))
pl.subplot(2,2,1)
pl.plot(lista,errorIn,'b',label="Training")
pl.plot(lista,errorOut,'c',label="Validation")
pl.legend(loc='upper left', bbox_to_anchor=(0, 0.5, 0.5, 0.5))
pl.xlabel(r"$\rho$")
pl.ylabel('Error Cuadratico Medio')
pl.title('EMC en función del valor de rho')

pl.subplot(2,2,2)
pl.plot(lista,varIn,'b',label="Training")
pl.plot(lista,varOut,'c',label="Validation")
pl.legend(loc='upper left', bbox_to_anchor=(0, 0.5, 0.5, 0.5))
pl.xlabel(r"$\rho$")
pl.ylabel('Varianza')
pl.title('Varianza en función del valor de rho')
pl.show()
```

```python
#Graficar los datos juntos a sus curva de regresion lineal con diferentes rho
RHO=[0,0.5,1,2,5,10]
pl.figure(figsize=(15,9))
pl.plot(x_inSample,y_inSample,'*b',label='testing')
pl.plot(x_outSample,y_outSample,'*c',label='validation')
pl.legend(loc='upper left', bbox_to_anchor=(0, 0.5, 0.5, 0.5))  
for rho in RHO:
    funcion=np.dot(x_inSample,reg_lineal(x_inSample,y_inSample,rho)[0])+reg_lineal(x_inSample,y_inSample,rho)[1]
    pl.plot(x_inSample,funcion,'-',label="rho="+str(rho))
pl.legend(loc='upper left', bbox_to_anchor=(0, 0.5, 0.5, 0.5))    
    
    
pl.xlabel('Average Area Income')
pl.ylabel('Price')
pl.title('Regresiones lineales con distintos rho')
pl.show()
```
